```python
clean_movies=movies.dropna()
```

```python
clean_movies.isnull().sum()
```

```python
movies.columns
```

```python
movies=clean_movies[['id','genres','keywords','original_title','overview']]
```

```python
movies
```

```python
movies = clean_movies[['id', 'genres', 'keywords', 'original_title', 'overview']].copy()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']
```

```python
movies.head()
```

```python
new_data=movies.drop(columns=['genres','overview','keywords'])
```

```python
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=1432,stop_words='english')
```

```python
cv
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
new_data[new_data['original_title']=="Spectre"].index[0]
```

```python
distance =sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])
for i in distance[0:5]:
   print(i)
```

```python
distance =sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])
for i in distance[0:5]:
   print(new_data.iloc[i[0]].original_title)
```

```python
def recommand(movies):
    index=new_data[new_data['original_title']==movies].index[0]
    distance =sorted(list(enumerate(similarity[index])),reverse=True, key=lambda x:x[1])
    for i in distance[0:5]:
      print(new_data.iloc[i[0]].original_title)
```

```python
recommand("Thor")
```

```python
recommand("Iron Man")
```
